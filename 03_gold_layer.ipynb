```python
import pandas as pd
import os

silver_path = "/home/jovyan/work/output/silver/"
gold_path = "/home/jovyan/work/output/gold/"

os.makedirs(gold_path, exist_ok=True)

orders = pd.read_csv(silver_path + "orders.csv")
items = pd.read_csv(silver_path + "order_items.csv")
customers = pd.read_csv(silver_path + "customers.csv")
products = pd.read_csv(silver_path + "products.csv")
sellers = pd.read_csv(silver_path + "sellers.csv")
payments = pd.read_csv(silver_path + "payments.csv")
```

```python
dim_customers = customers.copy()

dim_customers["customer_key"] = dim_customers["customer_id"].astype("category").cat.codes

dim_customers = dim_customers[[
    "customer_key",
    "customer_id",
    "customer_city",
    "customer_state"
]]
```

```python
dim_products = products.copy()

dim_products["product_key"] = dim_products["product_id"].astype("category").cat.codes

dim_products = dim_products[[
    "product_key",
    "product_id",
    "product_category_name"
]]
```

```python
dim_sellers = sellers.copy()

dim_sellers["seller_key"] = dim_sellers["seller_id"].astype("category").cat.codes

dim_sellers = dim_sellers[[
    "seller_key",
    "seller_id",
    "seller_city",
    "seller_state"
]]
```

```python
fact = items.merge(orders, on="order_id", how="left")
fact = fact.merge(payments, on="order_id", how="left")

# Keys
fact["customer_key"] = fact["customer_id"].astype("category").cat.codes
fact["product_key"] = fact["product_id"].astype("category").cat.codes
fact["seller_key"] = fact["seller_id"].astype("category").cat.codes

# Date
fact["order_date"] = pd.to_datetime(fact["order_purchase_timestamp"]).dt.date

# Handle null
fact["payment_value"] = fact["payment_value"].fillna(0)

fact = fact[[
    "order_id",
    "order_item_id",
    "customer_key",
    "product_key",
    "seller_key",
    "order_date",
    "price",
    "freight_value",
    "payment_value"
]]
```

```python
dim_customers.to_csv(gold_path + "dim_customers.csv", index=False)
dim_products.to_csv(gold_path + "dim_products.csv", index=False)
dim_sellers.to_csv(gold_path + "dim_sellers.csv", index=False)
fact.to_csv(gold_path + "fact_order_items.csv", index=False)

print("Gold layer created 🚀")
```
