```python
import pandas as pd
import os
```

```python
bronze_path = "/home/jovyan/work/output/bronze/"
silver_path = "/home/jovyan/work/output/silver/"

os.makedirs(silver_path, exist_ok=True)
```

```python
def clean_df(df, table_name):
    
    # Remove duplicates
    df = df.drop_duplicates()

    # Handle nulls
    for col in df.columns:
        df[col] = df[col].fillna("unknown")

    # Basic validation
    df["is_valid"] = True

    # Example validations (important 🔥)
    if "price" in df.columns:
        df.loc[df["price"] < 0, "is_valid"] = False

    if "order_purchase_timestamp" in df.columns and "order_delivered_customer_date" in df.columns:
        df.loc[df["order_delivered_customer_date"] < df["order_purchase_timestamp"], "is_valid"] = False

    return df
```

```python
tables = ["orders", "order_items", "customers", "products", "sellers", "payments"]

for table in tables:
    print(f"Processing {table}...")
    
    df = pd.read_csv(f"{bronze_path}{table}.csv")

    df = clean_df(df, table)

    df.to_csv(f"{silver_path}{table}.csv", index=False)

    print(f"Saved {table}")
```
